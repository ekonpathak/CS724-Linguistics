loading the sample audio to get number of samples and sample rate. 

```python
y,sr = librosa.load('hello.wav')
print("no of samples is:",len(y))
print("sample rate is means samples per second:",sr)
```

```python
print("time duration of audio is:",len(y)/sr)
```

time domain

```python
librosa.display.waveshow(y,x_axis='s')
```

# spectrogram 

```python
n_fft=1024 # window_length
```

```python
hop_length=320 #strides between the windows if step is smaller than window length ,the window will overlap
```

```python
s_scale=librosa.stft(y,n_fft=n_fft,hop_length=hop_length,win_length=n_fft)
```

```python
y_scale=np.abs(s_scale)**2 
print(y_scale.shape)
```

```python
librosa.display.specshow(y_scale,sr=sr,x_axis='time',y_axis='linear')
plt.title("Linear Frequency Power spectrogram")
plt.colorbar()
plt.tight_layout()
plt.show()
```

To make it more clear we convert frequency into mel scale

```python
mel_bins=64
librosa.display.specshow(y_scale,sr=sr,x_axis='time',y_axis='log')

plt.title("mel spectrogram")
plt.colorbar()
plt.tight_layout()
plt.show()
```

Transform the scale by taking log of amplitude into decible because human hear audio in terms of loudness

```python
y_log_scale=librosa.power_to_db(y_scale)
```

```python
librosa.display.specshow(y_log_scale,sr=sr,x_axis='time',y_axis='log')
plt.title("Log mel spectrogram")
plt.colorbar()
plt.tight_layout()
plt.show()
```

Mel filter banks

```python
filter_banks=librosa.filters.mel(n_fft=n_fft,sr=sr,n_mels=64)
filter_banks.shape
print("shape = [no of mel_bins x (frame_size/2)+1]")
```

```python
librosa.display.specshow(filter_banks,sr=sr,x_axis='linear')
plt.title('Mel filter Banks')
```

```python
import csv

#time = []
AX = []
AY =[]
AZ=[]

with open("test.csv", 'r') as datafile:
    plotting = csv.reader(datafile, delimiter=',')

    for ROWS in plotting:
        #time.append(float(ROWS[2]))
        AX.append(float(ROWS[0]))
        AY.append(float(ROWS[1]))
        AZ.append(float(ROWS[2]))

for i in range(len(AX)):
    if ((AX[i] >= min_standing_x and AX[i] <= max_standing_x) and (AY[i] >= min_standing_y and AY[i] <= max_standing_y) and (AZ[i] >= min_standing_z and AZ[i] <= max_standing_z)):
        print("AX: ",AX[i]," AY: ",AY[i]," AZ: ",AZ[i],"Standing")
    elif ((AX[i] >= min_sitting_x and AX[i] <= max_sitting_x) and (AY[i] >= min_sitting_y and AY[i] <= max_sitting_y) and (AZ[i] >= min_sitting_z and AZ[i] <= max_sitting_z)):
        print("AX: ",AX[i]," AY: ",AY[i]," AZ: ",AZ[i],"Sitting")
    elif ((AX[i] >= min_sleeping_x and AX[i] <= max_sleeping_x) and (AY[i] >= min_sleeping_y and AY[i] <= max_sleeping_y) and (AZ[i] >= min_sleeping_z and AZ[i] <= max_sleeping_z)):
        print("AX: ",AX[i]," AY: ",AY[i]," AZ: ",AZ[i],"Sleeping")
    else:
        print("AX: ",AX[i]," AY: ",AY[i]," AZ: ",AZ[i],"Pattern Not Recognized")
```
